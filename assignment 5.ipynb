```
#bài 1
def Max_Of_Two(a,b):
    return a if a>b else b
print(Max_Of_Two(*map(int,input().split())))
```

```
#bài 2
def swap(a,b):
    return b,a
a,b=map(int,input().split())
x,y=swap(a,b)
print(x,y)
```

```
#bài 3
def prime(n):
    if n<2:return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:return False
    return True
print(prime(int(input())))
```

```
#bài 4
def perfect(n):
    s=0
    for i in range(1,n):
        if n%i==0:s+=i
    return s==n
print(perfect(int(input())))
```

```
#bài 5
arr=list(map(int,input().split()))
k=int(input())
def find(arr,k):
    for i,v in enumerate(arr):
        if v==k:return i
    return -1
print(find(arr,k))
```

```
#bài 6
def fact(n):
    r=1
    for i in range(1,n+1): r*=i
    return r
print(fact(int(input())))
```

```
#bài 7
def calc(a,op,b):
    if op=='+': return round(a+b,2)
    if op=='-': return round(a-b,2)
    if op=='*': return round(a*b,2)
    return round(a/b,2)
a,op,b=input().split()
a=float(a); b=float(b)
print(f"{calc(a,op,b):.2f}")
```

```
#bài 8
def hamming(x,y):
    bx=bin(x)[2:]
    by=bin(y)[2:]
    L=max(len(bx),len(by))
    bx=bx.zfill(L)
    by=by.zfill(L)
    c=0
    for i in range(L):
        if bx[i]!=by[i]: c+=1
    return c
x,y=map(int,input().split())
print(hamming(x,y))
```

```
#bài 9
def sumDigits(n):
    s=0
    for c in str(int(n)):
        s+=int(c)
    return s
print(sumDigits(input()))
```

```
#bài 10
def iso(a,b):
    if len(a)!=len(b): return False
    m1={}
    m2={}
    for x,y in zip(a,b):
        if x in m1 and m1[x]!=y: return False
        if y in m2 and m2[y]!=x: return False
        m1[x]=y
        m2[y]=x
    return True
a,b=input().split()
print(str(iso(a,b)).lower())
```
