```python
#bài 10
n=int(input())
def col(x):
    c=1
    while x!=1:
        if x%2==0: x//=2
        else: x=3*x+1
        c+=1
    return c
best=1
bestL=1
for i in range(1,n+1):
    L=col(i)
    if L>bestL:
        best=i
        bestL=L
print(best,bestL)
```
